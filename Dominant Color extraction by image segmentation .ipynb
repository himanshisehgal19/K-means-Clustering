# Dominant Color Extraction for Image Segmentation


```python
# -Image Segmentation
# -Segmentation partitions an image into regions having similar visual appearance corresponding to parts of objects
# -We will try to extract the most dominant 'K' Colors using K-Means
# -We can apply K-Means with each pixel will reassigned to the closest of the K Colors, leading to segmentation
```

```python
import matplotlib.pyplot as plt
import cv2
```

```python
im=cv2.imread('elephant.jpg') #reads an image into BGR format
#img=cv2.imread('elephant.jpg')
im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB) #converts the image to RGB so that it can show exact image
orignalshape=im.shape
```

```python
plt.imshow(im)
```

```python
#plt.imshow(img)
```

```python
allpixels=im.reshape(-1,3)
allpixls=im.reshape(-1,3)
```

```python
allpx=im.reshape(-1,3)
```

```python
import numpy as np
from sklearn.cluster import KMeans
```

```python
km2=KMeans(n_clusters=2)
km2.fit(allpx)
```

```python
dominant_colors=4
km=KMeans(n_clusters=dominant_colors)
kmx=KMeans(n_clusters=8)
km.fit(allpixels)
```

```python
kmx.fit(allpixls)
```

```python
centrs=kmx.cluster_centers_
```

```python
centers=km.cluster_centers_
```

```python
cen=km2.cluster_centers_
```

```python
centers=np.array(centers,dtype='uint8')
centrs=np.array(centrs,dtype='uint8')
cen=np.array(cen,dtype='uint8')
```

```python
print(centers)
```

```python
print(cen)
```

```python
print(centrs)
```

```python
i=1
plt.figure(0,figsize=(8,2))
colors=[]
for each_col in centers:
    plt.subplot(1,4,i)
    plt.axis("off")
    i+=1
    colors.append(each_col)
    a=np.zeros((100,100,3),dtype='uint8')
    a[:,:,:]=each_col
    plt.imshow(a)
```

```python
j=1
plt.figure(0,figsize=(8,2))
color=[]
for each_cl in centrs:
    plt.subplot(1,8,j)
    plt.axis("off")
    j+=1
    color.append(each_cl)
    a=np.zeros((100,100,3),dtype='uint8')
    a[:,:,:]=each_cl
    plt.imshow(a)
```

```python
i=1
plt.figure(0,figsize=(8,2))
clr=[]
for each in cen:
    plt.subplot(1,2,i)
    plt.axis("off")
    i+=1
    clr.append(each)
    a=np.zeros((100,100,3),dtype='uint8')
    a[:,:,:]=each
    plt.imshow(a)
```

```python
newimg2=np.zeros((330*500,3),dtype='uint8')
km2.labels_
for ix in range(newimg2.shape[0]):
    newimg2[ix]=clr[km2.labels_[ix]]
```

```python
newimg2=newimg2.reshape((orignalshape))
plt.imshow(newimg2)
plt.show()
```

```python
new_img=np.zeros((330*500,3),dtype='uint8')
print(new_img.shape)
```

```python
newimgN=np.zeros((330*500,3),dtype='uint8')
print(newimgN.shape)
```
